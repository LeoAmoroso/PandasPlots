## Exercício 2

**a)** Utilizando a base de dados de covid, crie um gráfico de linhas mostrando a evolução do número de casos novos mês a mês, para as DRS 'Grande São Paulo', 'São José do Rio Preto' e 'Marília'. Identifique os DRS por cores e legenda. Não esqueça de colocar título para o gráfico e para os eixos. 

```python
table = df2.pivot_table(index='ano_mes', columns='nome_drs', values='casos_novos', aggfunc='sum')
table = table[['Grande São Paulo', 'Marília', 'São José do Rio Preto']]
table
```

```python
table.plot(xlabel = 'Meses', ylabel = 'Numero de casos', title='Casos Novos', figsize=(10,5))
```

**b)** Repita o gráfico do item a trocando as cores dos DRS.

```python
table.plot(xlabel = 'Meses', ylabel = 'Numero de casos', title='Casos Novos', figsize=(10,5), color=['#936639','#ddb892','#9b2226'])
```

**c)** Considerando apenas os DRS 'Bauru', 'São José do Rio Preto' e 'Marília', faça três gráficos de pizza, para os meses de abril, maio e junho, mostrando a porcentagem de novos óbitos em cada DRS.

```python
table2 = df2.pivot_table(index='ano_mes', columns='nome_drs', values='obitos_novos', aggfunc='sum')
table2 = table2[['Bauru', 'São José do Rio Preto', 'Marília']]
table2.plot(kind='pie', subplots=True, figsize=(20,10), labeldistance=None, autopct='%1.0f%%', title='New Deaths')
```
